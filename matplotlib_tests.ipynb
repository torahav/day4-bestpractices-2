```python
import numpy as np

labels = ['Bound', 'Sampling', 'Free']
wt_occ = [50, 35, 15]
mut_occ = [10, 2, 88]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(1)
rects1 = ax.bar(x - width/2, wt_occ, width, label='WT')
rects2 = ax.bar(x + width/2, mut_occ, width, label='Mutant')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('partial occupancy')
ax.set_title('State occupancy WT vs mutant TF-HaloTag')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

#fig.tight_layout()

plt.show()
```

```python
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('./monster.jpg')

imgplot = plt.imshow(img)
```

```python
# Trying to convert RGB to greyscale using formula Y' = 0.2989R + 0.5870G + 0.1140B.
# Seems like it makes more sense to use Pillow library for images rather than matplotlib

def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989*r + 0.5870*g + 0.1140*b
    return gray

gray = rgb2gray(img)    
plt.imshow(gray, cmap=plt.get_cmap('gray'))
plt.show()
```

```python
import matplotlib.patches as patches

fig, ax = plt.subplots()
im = ax.imshow(img)
patch = patches.Circle((1150, 1000), radius=1000, transform=ax.transData)
im.set_clip_path(patch)

ax.axis('off')
plt.show()
```
